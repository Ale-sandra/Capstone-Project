```python
#Esraggo i singoli file degli anni dalla cartella di lavoro DATASET RINNOVI
import pandas as pd
import os

#Percorso del file
Percorso = r"C:\Users\aless\Desktop\PROGETTO FINALE - AGENZIA"
file_path = os.path.join(Percorso, "DATASET RINNOVI.xlsx")

#Leggi tutti i fogli
all_sheets = pd.read_excel(file_path, sheet_name=None)

#Cicla e salva ogni foglio
for nome_foglio, df in all_sheets.items():
    output_file = os.path.join(Percorso, f"Dettaglio_movimenti_RCA_{nome_foglio}.xlsx")
    df.to_excel(output_file, index=False)
    print(f" Creato: {output_file}")

```
